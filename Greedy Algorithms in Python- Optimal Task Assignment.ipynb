## Greedy Algorithms in Python: Optimal Task Assignment

I am going to be solving the optimal task assignment problem and in order to do that I am  going to make use of a greedy algorithm strategy. 

So what we'll be doing is we'll be defining what this optimal task assignment problem is we'll be outlining how we can solve this problem using a
greedy approach and then we'll code up that solution in Python 

Definition of optimal task assignment problem:  given an array and each cell of the array contains a number and that number
corresponds to the duration of a task. 

For instance we can think about the numbers let's say corresponding to hours. So the task  in the first cell corresponds to taking six hours, and the next  is three hours.

We want to assign the tasks  to some workers and we want to determine the optimal way in
which to assign these tasks to the workers assuming that: 
1. Each worker must take two tasks 
2. The tasks are independent 
3. Each task does take a fixed amount of time which is given by the number in this array

The problem that we want to solve is we want to assign the tasks to the workers such that the time it takes
complete all of the tasks is minimized.



We want something that kind of looks like above,  where we assign each worker a pair of
tasks that take a certain amount of time so we want to assign the tasks the workers such that the time it takes to
complete all the tasks together is minimized







This is the optimal solution!

To think about how to solve this problem in a more general way. 

For given an array of arbitrary length with some durations: how
can we solve this problem arbitrarily. 

One possible way that we could solve it is by enumerated all the possible pairs that we could make from this array.That is, going through the array and making every single pair  to minimized the hours.






This is the  brute-force way;  if we enumerate all the possible pairs for a given list let's say our arrays of size n

This doesn't scale too well. The other approach is to use a greedy approach and the one that happens to work in this case is the
following where we'll pair the longest task with the shortest one and we just keep doing this until we run out of tasks in the array that is given

and then we do the same approach as we go through the array so now we pick the next biggest one and the next smallest one
